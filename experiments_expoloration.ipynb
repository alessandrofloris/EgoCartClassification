## Custom CNN

```python
import json
```

```python
customcnn_history_path = "experiments/customcnn/"
history_path_1 = customcnn_history_path + "exp_1/train_customcnn.json"
history_path_2 = customcnn_history_path + "exp_2/train_customcnn.json"
history_path_3 = customcnn_history_path + "exp_3/train_customcnn.json"
history_path_4 = customcnn_history_path + "exp_4/train_customcnn.json"
history_path_5 = customcnn_history_path + "exp_5/train_customcnn.json"
```

```python
import matplotlib.pyplot as plt

with open(history_path_1, "r") as f:
    history = json.load(f)

epochs = [e["epoch"] for e in history["epochs"]]
train_loss = [e["train_loss"] for e in history["epochs"]]
val_loss = [e["val_loss"] for e in history["epochs"]]

fig = plt.figure(figsize=(10, 5))
plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, val_loss, label="Val Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)
plt.show()
fig.savefig('loss_curves_exp_1.png', dpi=300)
```
